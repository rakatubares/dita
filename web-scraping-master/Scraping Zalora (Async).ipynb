```python
import urllib

URL_PRODUK = "http://www.zalora.co.id/_c/rpc?&req={%22method%22%3A%22Costa.ListCatalogProducts"
URL_PRODUK += "%22%2C%22params%22%3A[{%22category_id%22%3A[%22%22]%2C%22limit%22%3A72%2C%22offset%22%3A360"
URL_PRODUK += "%2C%22segment%22%3A%22women%22%2C%22gender%22%3A[%22women%22]%2C%22dir%22%3A%22desc%22%2C%22sort%22%3A"
URL_PRODUK += "%22popularity%22%2C%22all_products%22%3Atrue%2C%22catalog_type%22%3A%22%22%2C%22campaign%22%3A%22cmpgn_five"
URL_PRODUK += "%3Abranded-sale%22%2C%22url_key%22%3A%22%2Fwomen%2Fsemua-produk%22}]}&lang=id"

r = urllib.urlopen(URL_PRODUK).read()
print(type(r))
print(r[1:100])
```

```python
import json

prods = json.loads(str(r))
print(type(prods))
for key, value in prods['result']['response']['docs'].iteritems():
    print(key)
```
